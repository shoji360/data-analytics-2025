# クラウドのデータベース「BigQuery」と接続
- BigQueryとは･･･
  - Googleの提供しているオンラインで使える分析プラットフォーム
  - 様々な機能があるが、SQLに対応しており、単純なDBとしても使える
  - 従量課金制
  - 詳しく知りたい人はWikipediaへ
    - https://ja.wikipedia.org/wiki/BigQuery
- 今回は、授業で使うデータの入ったDBを、あらかじめ用意しておきました！（山本先生が）

- 左側ツールバーの「INTEGRATIONS」の「＋」をクリックし･･･
image.png
- 「BigQuery」を選択
